# BLAST in BioPython

## Context:
- BLAST - Basic Local Alignment Search Tool
- Enables to search query molecule with a library/database of sequences
- E-value: number of expected hits of similar quality (score) that could be found just by chance.

```python
from Bio.Blast import NCBIWWW
help(NCBIWWW.qblast)
```

```python
result_handle = NCBIWWW.qblast("blastn", "nt", "8332116")
```

```python
# BLAST generates output in XML format and we need to somehow parse it!
from Bio.Blast import NCBIXML
blast_records = NCBIXML.parse(result_handle)
blast_record = next(blast_records)
```

```python
# blast_records = list(blast_records)
# blast_records
```

```python
E_VALUE_THRESH = 0.04
count = 0
for alignment in blast_record.alignments:
    for hsp in alignment.hsps:
        if hsp.expect < E_VALUE_THRESH:
            count += 1
            print("****Alignment****")
            print("sequence:", alignment.title)
            print("length:", alignment.length)
            print(hsp.query[0:75] + "...")
            print(hsp.match[0:75] + "...")
            print(hsp.sbjct[0:75] + "...")
            print()
                
print(f"There are {count} similar sequences in Blast output")
```

Interpretation: The list shows the resulting nucleotide sequences which are found to be similar (threshold=0.04) with our query sequence.

## Summary: In this tutorial, we run the BLAST search, parsed the result, and printed the resulting nucleotide sequences which are found to be similar (threshold=0.04) with our query sequence.
